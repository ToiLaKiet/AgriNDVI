```python
import os
import pandas as pd

# List to store DataFrames
dataframes = []

# Iterate through files in the current directory
for file in os.listdir('.'):
    if file.endswith('.csv'):  # Check if the file is a CSV
        df = pd.read_csv(file)  # Read the CSV file
        dataframes.append(df)   # Append the DataFrame to the list

# Concatenate all DataFrames
combined_df = pd.concat(dataframes, ignore_index=True)

# Sort the combined DataFrame by the 'date' column
combined_df = combined_df.sort_values(by='date')

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('final_2016_2025_ERA5.csv', index=False)

print("All CSV files have been concatenated and sorted by 'date' into 'final_2016_2025_ERA5.csv'.")
```
