```python
import pandas as pd

# Load the CSV file
csv_file = "prediction_aligned_window_based_engineered_features.csv"
data = pd.read_csv(csv_file)

# Ensure 'start_date' is datetime
data['start_date'] = pd.to_datetime(data['start_date'], errors='coerce')

# --- Split theo năm ---
TrainSet = data[data['start_date'].dt.year < 2024]      # Train: trước 2024
TestSet  = data[data['start_date'].dt.year == 2024]     # Test: năm 2024
ValSet   = data[data['start_date'].dt.year == 2025]     # Validation: năm 2025

# Print sizes
print("Train samples :", len(TrainSet))
print("Test samples  :", len(TestSet))
print("Val samples   :", len(ValSet))

# --- Save to CSV ---
TrainSet.to_csv("TrainSet.csv", index=False)
TestSet.to_csv("TestSet_2024.csv", index=False)
ValSet.to_csv("ValSet_2025.csv", index=False)

print("Done! Files saved:")
print(" - TrainSet.csv")
print(" - TestSet_2024.csv")
print(" - ValSet_2025.csv")
```
