```python
# ===== Kiểm tra cột tối thiểu =====
REQUIRED_ERA5 = [
    "NAME_1","GID_1","NAME_2","GID_2","date",
    "2m_temperature_mean",
    "volumetric_soil_water_layer_1_mean",
    "volumetric_soil_water_layer_2_mean",
    "total_precipitation_sum",                 # mm/day (bạn xác nhận là mm)
    "surface_solar_radiation_downwards_sum",
]
miss = [c for c in REQUIRED_ERA5 if c not in df.columns]
if miss:
    raise ValueError(f"Thiếu cột ERA5 tối thiểu: {miss}")
if not {"GID_2","start_date","end_date"}.issubset(ndvi.columns):
    raise ValueError("NDVI cần có cột 'GID_2','start_date','end_date'.")
```

```python
# ===== Chuẩn hóa =====
df = df.copy()
ndvi = ndvi.copy()
df["date"] = pd.to_datetime(df["date"], errors="coerce")
ndvi["start_date"] = pd.to_datetime(ndvi["start_date"], errors="coerce")
ndvi["end_date"]   = pd.to_datetime(ndvi["end_date"], errors="coerce")

df = df.sort_values(["GID_2","date"]).reset_index(drop=True)
ndvi = ndvi.sort_values(["GID_2","start_date"]).reset_index(drop=True)
```

```python
# ===== Cấu hình ====
WIN = 16                      # MOD13Q1 là composite 16 ngày
RAIN_MM_THRESHOLD = 1.0       # ngưỡng ngày-mưa (mm/day)
STRICT_DTR = False            # True nếu có 2m_temperature_max/min (daily)
REQUIRE_FULL_WINDOW = True    # True: yêu cầu đủ đúng 16 ngày; False: chấp nhận cửa sổ ngắn
MIN_DAYS = 12                 # dùng khi REQUIRE_FULL_WINDOW=False
WINDY_THRESH = 6.0            # m/s, ngưỡng đếm "ngày gió mạnh"
```

```python
# ===== Các cột tùy chọn (nếu có thì tính thêm) =====
HAS_TMAX = "2m_temperature_max" in df.columns
HAS_TMIN = "2m_temperature_min" in df.columns
HAS_ET   = "total_evaporation_sum" in df.columns
HAS_PET  = "potential_evaporation_sum" in df.columns
HAS_NETS = "surface_net_solar_radiation_sum" in df.columns
HAS_WIND = {"10m_u_component_of_wind_mean","10m_v_component_of_wind_mean"}.issubset(df.columns)
HAS_PRES = "surface_pressure_mean" in df.columns

if STRICT_DTR and not (HAS_TMAX and HAS_TMIN):
    raise ValueError("STRICT_DTR=True nhưng thiếu 2m_temperature_max/min.")
```

```python
# ===== Ngưỡng nắng p75 theo huyện (climatology toàn chuỗi) =====
solar_p75_by_gid2 = (
    df.groupby("GID_2")["surface_solar_radiation_downwards_sum"]
      .quantile(0.75)
      .rename("solar_p75_gid2")
)
df = df.merge(solar_p75_by_gid2, on="GID_2", how="left")
```

```python
# ===== Helpers =====
def run_lengths(mask_bool: np.ndarray):
    """Độ dài các chuỗi True liên tiếp."""
    if mask_bool.size == 0:
        return []
    m = mask_bool.astype(np.int8)
    edges = np.diff(np.concatenate(([0], m, [0])))
    starts = np.where(edges == 1)[0]
    ends   = np.where(edges == -1)[0]
    return (ends - starts).tolist()

def nan_slope(y: np.ndarray):
    """Hệ số dốc tuyến tính (OLS) của chuỗi theo chỉ số 0..n-1, bỏ NaN; trả NaN nếu <2 điểm."""
    y = np.asarray(y, dtype=float)
    idx = np.arange(len(y))
    mask = ~np.isnan(y)
    if mask.sum() < 2:
        return np.nan
    x = idx[mask].astype(float)
    yy = y[mask].astype(float)
    x_mean = x.mean(); y_mean = yy.mean()
    denom = ((x - x_mean)**2).sum()
    if denom == 0:
        return np.nan
    return ((x - x_mean)*(yy - y_mean)).sum() / denom

def to_mm_positive(arr_like):
    """
    Chuẩn hoá ET/PET:
    - Nếu dấu âm (flux lên), đảo dấu về dương.
    """
    x = np.asarray(arr_like, dtype=float)
    if np.nanmedian(x) < 0:
        x = -x
    return x
```

```python
# ===== Hàm tính feature cho 1 khoảng NDVI =====
def summarize_window(w: pd.DataFrame, strict_dtr: bool):
    # vector cơ bản
    precip = w["total_precipitation_sum"].to_numpy()       # mm/day (đã xác nhận)
    tmean  = w["2m_temperature_mean"].to_numpy()
    sm1    = w["volumetric_soil_water_layer_1_mean"].to_numpy()
    sm2    = w["volumetric_soil_water_layer_2_mean"].to_numpy()
    sw_dn  = w["surface_solar_radiation_downwards_sum"].to_numpy()
    sw_p75 = w["solar_p75_gid2"].to_numpy()
    sunny  = sw_dn > sw_p75
    wet    = precip > RAIN_MM_THRESHOLD

    # ==== (A) Các đặc trưng "chính" ====
    rainy_days16      = int(np.nansum(wet))
    sunny_days16      = int(np.nansum(sunny))
    p95_precip1d_16   = float(np.nanpercentile(precip, 95))
    max_sw_dn1d_16    = float(np.nanmax(sw_dn))
    sm_diff16         = float(np.nanmean(sm1 - sm2))

    if strict_dtr:
        tmax_d = w["2m_temperature_max"].to_numpy()
        tmin_d = w["2m_temperature_min"].to_numpy()
        tmax16_mean = float(np.nanmean(tmax_d))
        tmin16_mean = float(np.nanmean(tmin_d))
        dtr16       = float(np.nanmean(tmax_d - tmin_d))
        dtr_is_proxy = False
    else:
        # proxy từ 2m_temperature_mean trong khung
        tmax16_mean = float(np.nanpercentile(tmean, 90)) # Âm cực đại 10% là xấp xỉ tmax thực 
        tmin16_mean = float(np.nanpercentile(tmean, 10)) # Âm cực tiểu 10% là xấp xỉ tmin thực
        dtr16       = float(tmax16_mean - tmin16_mean)
        dtr_is_proxy = True

    # ==== (B) WINDOW FEATURES THÔNG DỤNG ====
    # 1) Precip summary
    precip_sum16   = float(np.nansum(precip))
    precip_mean16  = float(np.nanmean(precip))
    precip_std16   = float(np.nanstd(precip, ddof=0))
    precip_p90_16  = float(np.nanpercentile(precip, 90))
    precip_p10_16  = float(np.nanpercentile(precip, 10))
    wet_frac16     = float(rainy_days16 / len(w)) if len(w) > 0 else np.nan

    wet_runs = run_lengths(wet)
    dry_runs = run_lengths(~wet)
    wet_spell_max16 = int(max(wet_runs) if wet_runs else 0)
    dry_spell_max16 = int(max(dry_runs) if dry_runs else 0)

    # 2) Solar summary
    sw_dn_sum16  = float(np.nansum(sw_dn))
    sw_dn_mean16 = float(np.nanmean(sw_dn))
    sw_dn_p90_16 = float(np.nanpercentile(sw_dn, 90))
    if HAS_NETS:
        sw_net = w["surface_net_solar_radiation_sum"].to_numpy()
        sw_net_sum16  = float(np.nansum(sw_net))
        sw_net_mean16 = float(np.nanmean(sw_net))
    else:
        sw_net_sum16  = np.nan
        sw_net_mean16 = np.nan

    # 3) Temperature (mean) summary & trend
    tmean16_mean  = float(np.nanmean(tmean))
    tmean16_std   = float(np.nanstd(tmean, ddof=0))
    tmean_p90_16  = float(np.nanpercentile(tmean, 90))
    tmean_p10_16  = float(np.nanpercentile(tmean, 10))
    tmean_trend16 = float(nan_slope(tmean))  # °C/ngày. nan_slope là hàm tự định nghĩa ở trên

    # 4) Soil moisture summary
    sm1_mean16     = float(np.nanmean(sm1))
    sm2_mean16     = float(np.nanmean(sm2))
    sm_diff_std16  = float(np.nanstd(sm1 - sm2, ddof=0))

    # 5) Evap/PET (nếu có)
    if HAS_ET:
        et_mm = to_mm_positive(w["total_evaporation_sum"].to_numpy())
        et16_sum = float(np.nansum(et_mm))
        et16_mean = float(np.nanmean(et_mm))
    else:
        et16_sum = np.nan
        et16_mean = np.nan

    if HAS_PET:
        pet_mm = to_mm_positive(w["potential_evaporation_sum"].to_numpy())
        pet16_sum = float(np.nansum(pet_mm))
        pet16_mean = float(np.nanmean(pet_mm))
    else:
        pet16_sum = np.nan
        pet16_mean = np.nan

    if HAS_ET and HAS_PET:
        evap_deficit16 = float(np.nansum(np.maximum(pet_mm - et_mm, 0.0)))
        et_pet_ratio16 = float(et16_sum / pet16_sum) if pet16_sum and pet16_sum != 0 else np.nan
    else:
        evap_deficit16 = np.nan
        et_pet_ratio16 = np.nan

    # 6) Wind speed (nếu có u,v)
    if HAS_WIND:
        u = w["10m_u_component_of_wind_mean"].to_numpy()
        v = w["10m_v_component_of_wind_mean"].to_numpy()
        win_speed = np.hypot(u, v)                        # m/s
        wind_speed_mean16 = float(np.nanmean(win_speed))
        wind_speed_max16  = float(np.nanmax(win_speed))
        wind_speed_p90_16 = float(np.nanpercentile(win_speed, 90))
        windy_days16      = int(np.nansum(win_speed > WINDY_THRESH))
    else:
        wind_speed_mean16 = np.nan
        wind_speed_max16  = np.nan
        wind_speed_p90_16 = np.nan
        windy_days16      = np.nan

    # 7) Surface pressure (nếu có)
    if HAS_PRES:
        sp = w["surface_pressure_mean"].to_numpy()
        sp_mean16 = float(np.nanmean(sp))
        sp_std16  = float(np.nanstd(sp, ddof=0))
    else:
        sp_mean16 = np.nan
        sp_std16  = np.nan

    return {
        # === “chính” ===
        "rainy_days16": rainy_days16,
        "sunny_days16": sunny_days16,
        "p95_precip1d_16": p95_precip1d_16,
        "max_sw_dn1d_16": max_sw_dn1d_16,
        "tmax16_mean": tmax16_mean,
        "tmin16_mean": tmin16_mean,
        "dtr16": dtr16,
        "sm_diff16": sm_diff16,
        "dtr_is_proxy": dtr_is_proxy,
        # === window features phổ biến ===
        # Precip
        "precip_sum16": precip_sum16,
        "precip_mean16": precip_mean16,
        "precip_std16": precip_std16,
        "precip_p90_16": precip_p90_16,
        "precip_p10_16": precip_p10_16,
        "wet_frac16": wet_frac16,
        "wet_spell_max16": wet_spell_max16,
        "dry_spell_max16": dry_spell_max16,
        # Solar
        "sw_dn_sum16": sw_dn_sum16,
        "sw_dn_mean16": sw_dn_mean16,
        "sw_dn_p90_16": sw_dn_p90_16,
        "sw_net_sum16": sw_net_sum16,
        "sw_net_mean16": sw_net_mean16,
        # Temperature mean & trend
        "tmean16_mean": tmean16_mean,
        "tmean16_std": tmean16_std,
        "tmean_p90_16": tmean_p90_16,
        "tmean_p10_16": tmean_p10_16,
        "tmean_trend16": tmean_trend16,
        # Soil moisture
        "sm1_mean16": sm1_mean16,
        "sm2_mean16": sm2_mean16,
        "sm_diff_std16": sm_diff_std16,
        # Evap/PET
        "et16_sum": et16_sum,
        "et16_mean": et16_mean,
        "pet16_sum": pet16_sum,
        "pet16_mean": pet16_mean,
        "evap_deficit16": evap_deficit16,
        "et_pet_ratio16": et_pet_ratio16,
        # Wind speed
        "wind_speed_mean16": wind_speed_mean16,
        "wind_speed_max16":  wind_speed_max16,
        "wind_speed_p90_16": wind_speed_p90_16,
        "windy_days16":      windy_days16,
        # Surface pressure
        "sp_mean16": sp_mean16,
        "sp_std16":  sp_std16,
    }
```

```python
# ===== Tính features theo từng khoảng ngày của bộ NDVI =====
out_rows = []
for gid2, g in df.groupby("GID_2"):
    ndvi_g = ndvi[ndvi["GID_2"] == gid2] # Danh sách các khoảng NDVI cho GID_2 này
    if ndvi_g.empty: # không có khoảng NDVI nào
        continue

    g = g.sort_values("date")
    for _, r in ndvi_g.iterrows():
        s = r["start_date"]; e = r["end_date"]
        if pd.isna(s) or pd.isna(e): # bỏ nếu ngày không hợp lệ
            continue

        w = g[(g["date"] >= s) & (g["date"] <= e)]
        if REQUIRE_FULL_WINDOW:
            if len(w) != WIN:
                continue
        else:
            if len(w) < MIN_DAYS:
                continue

        feat = summarize_window(w, STRICT_DTR)
        out_rows.append({
            "NAME_1": w.iloc[-1]["NAME_1"],
            "GID_1":  w.iloc[-1]["GID_1"],
            "NAME_2": w.iloc[-1]["NAME_2"],
            "GID_2":  w.iloc[-1]["GID_2"],
            "start_date": s,
            "end_date": e,
            # giữ lại các cột khác của NDVI (nếu có)
            **({k: r[k] for k in ndvi.columns if k not in ["GID_2","start_date","end_date"]}),
            **feat
        })

features16_aligned = (
    pd.DataFrame(out_rows)
      .sort_values(["GID_2","start_date"])
      .reset_index(drop=True)
)

print(features16_aligned.head())
# features16_aligned.to_csv("era5_features16_aligned_to_MOD13Q1_plus_wind.csv", index=False)
```
