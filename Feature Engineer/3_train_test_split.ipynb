```python
import pandas as pd

# Load the CSV file
csv_file = "prediction_aligned_window_based_engineered_features.csv"  # Replace with your CSV file name
data = pd.read_csv(csv_file)

# Ensure 'start_date' is in datetime format
data['start_date'] = pd.to_datetime(data['start_date'], errors='coerce')

# Split the data into two based on the year
TestSet = data[data['start_date'].dt.year >= 2025]
TrainingSet = data[data['start_date'].dt.year < 2025]

# Save the split data into two separate CSV files
TestSet.to_csv("TestSet.csv", index=False)
TrainingSet.to_csv("TrainingSet.csv", index=False)
```
