```python
import pandas as pd 
import numpy as np
```

```python
df_train = pd.read_csv('TrainingSet.csv',parse_dates=['start_date', 'end_date'])
df_test = pd.read_csv('TestSet.csv',parse_dates=['start_date', 'end_date'])
```

```python
df = df_test.copy()
```

# **I . Sin/Cos Date Encoding**

```python
df['middle_date'] = df['start_date'] + (df['end_date'] - df['start_date']) / 2
```

```python
# Extract year
df['year'] = df['middle_date'].dt.year

# Compute day-of-year within each year
df['day_of_year'] = df.groupby('year')['middle_date'].transform(lambda x: x.dt.dayofyear)

# Normalize day in each year
df['day_norm'] = df['day_of_year'] / df.groupby('year')['day_of_year'].transform('max')

# Compute cyclical features
df['date_sin'] = np.sin(2 * np.pi * df['day_norm'])
df['date_cos'] = np.cos(2 * np.pi * df['day_norm'])
```

```python
df.drop(columns=['end_date', 'start_date'], inplace=True)
df
```

# **II. Time lagged NDVI features**

```python
# ---- cấu hình ----
k = 3  # dùng 3 bước thời gian quá khứ: t-1..t-3
group_col = 'GID_2' if 'GID_2' in df.columns else None
```

```python
# tìm cột ngày
_date_candidates = ['middle_date', 'date', 'Date', 'DATE', 'ds']
date_col = next((c for c in _date_candidates if c in df.columns), None)
if date_col is None:
    raise ValueError("Không tìm được cột ngày (middle_date/date/Date/...) trong df_train")
```

```python
# ---- chuẩn hóa ngày & sort ----
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
```

```python
# sort theo vùng + thời gian (year chỉ để rõ ràng, date đã chứa year rồi)
sort_cols = [c for c in [group_col, 'year', date_col] if c is not None]
df = df.sort_values(sort_cols)
```

```python
# helper: group theo vùng (KHÔNG group theo year để nối qua ranh năm)
def gseries(col):
    return df.groupby(group_col)[col] if group_col else df[col]
```

```python
# ---- (1) Lag features cho ndvi_mean ----
for L in range(1, k + 1):
    if group_col:
        df[f'ndvi_mean_lag{L}'] = gseries('ndvi_mean').shift(L) # shift về quá khứ L bước
    else:
        df[f'ndvi_mean_lag{L}'] = df['ndvi_mean'].shift(L)
```

```python
df
```

```python
# ---- (2) Rolling (t-1..t-k) trên ndvi_mean ----
if group_col:
    df[f'ndvi_mean_roll{k}_mean'] = gseries('ndvi_mean').transform(
        lambda s: s.shift(1).rolling(window=k, min_periods=1).mean() # shift(1) để không tính cả giá trị hiện tại. window = k là k bước trước đó
    )
    df[f'ndvi_mean_roll{k}_std'] = gseries('ndvi_mean').transform(
        lambda s: s.shift(1).rolling(window=k, min_periods=1).std() 
    )
    df[f'ndvi_mean_roll{k}_min'] = gseries('ndvi_mean').transform(
        lambda s: s.shift(1).rolling(window=k, min_periods=1).min()
    )
    df[f'ndvi_mean_roll{k}_max'] = gseries('ndvi_mean').transform(
        lambda s: s.shift(1).rolling(window=k, min_periods=1).max()
    )
else:
    s = df['ndvi_mean'].shift(1)
    df[f'ndvi_mean_roll{k}_mean'] = s.rolling(k, min_periods=1).mean()
    df[f'ndvi_mean_roll{k}_std']  = s.rolling(k, min_periods=1).std()
    df[f'ndvi_mean_roll{k}_min']  = s.rolling(k, min_periods=1).min()
    df[f'ndvi_mean_roll{k}_max']  = s.rolling(k, min_periods=1).max()

df
```

```python
# ---- (3) Expanding (từ đầu chuỗi đến t-1) ----
if group_col:
    df['ndvi_mean_exp_mean'] = gseries('ndvi_mean').transform(
        lambda s: s.shift(1).expanding(min_periods=1).mean()
    )
    df['ndvi_mean_exp_std'] = gseries('ndvi_mean').transform(
        lambda s: s.shift(1).expanding(min_periods=1).std()
    )
else:
    s = df['ndvi_mean'].shift(1)
    df['ndvi_mean_exp_mean'] = s.expanding(min_periods=1).mean()
    df['ndvi_mean_exp_std']  = s.expanding(min_periods=1).std()

df

```
